```python
import pandas as pd
raw_data = pd.read_json("../../sciso.jsonl", lines=True)
```

```python
def merge_concepts(series):
    merged_concepts = {}
    for concepts in series:
        try:
            for c in concepts:
                if (
                    c["display_name"] not in merged_concepts
                    or merged_concepts[c["display_name"]]["score"] < c["score"]
                ):
                    merged_concepts[c["display_name"]] = {
                        "name": c["display_name"],
                        "score": c["score"],
                        "level": c["level"],
                    }
        except Exception:
            print(concepts)
    return merged_concepts


def group_by_title(df: pd.DataFrame):
    retval = pd.DataFrame(list(df["metadata"]))
    retval = retval[
        [
            "title",
            "normalized_venue",
            "year",
            "citation_count",
            "authors",
            "type",
            "open_access",
            "external_ids",
            "concepts",
        ]
    ].join(df[["PostId", "AddedAt"]])
    retval["year"] = retval["year"].astype("Int64")
    retval = (
        retval.groupby("title")
        .agg(
            {
                "normalized_venue": lambda x: set(x),
                "year": lambda x: set(x),
                "citation_count": "max",
                "open_access": "any",
                "PostId": lambda x: list(x),
                "concepts": merge_concepts,
                "AddedAt": list,
            }
        )
        .reset_index()
    )
    retval["concepts"] = retval["concepts"].apply(
        lambda x: sorted(x.values(), key=lambda y: y["score"], reverse=True)
    )
    return retval


grouped_data = group_by_title(raw_data)
grouped_data.head()
```

```python
from pymongo import MongoClient

db = MongoClient("localhost", 27017).get_database("sciso-2024")
pub_refs = db.get_collection("pub_refs_normalized_union")


def resolve_conflict(resolver):
    for solution in resolver.items():
        idx, changes = solution
        for post in grouped_data.loc[idx]["PostId"]:
            cursor = pub_refs.find({"PostId": str(post)})
            for doc in cursor:
                if grouped_data.loc[idx]["title"] != doc["metadata"]["title"]:
                    print("Title mismatch", idx, doc["metadata"]["title"])
                    continue
                pub_refs.update_one({"_id": doc["_id"]}, {"$set": changes})
                pub_refs.update_one(
                    {"_id": doc["_id"]},
                    {"$unset": {"normalized_venue": "", "concepts": ""}},
                )
                print(doc["PostId"], "updated")

# resolve_conflict(resolve_venue_conflicts)
```
